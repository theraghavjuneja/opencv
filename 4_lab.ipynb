```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def analyze_histogram(image_path):
    """
    Computes and analyzes the histogram values for grayscale and color images.

    Args:
        image_path: Path to the image file.
    """

    try:
        # Read the image
        img = cv2.imread(image_path)

        if img is None:
            raise ValueError(f"Image not found at {image_path}")

        color_histograms = []
        color_labels = ['Blue', 'Green', 'Red']

        plt.figure(figsize=(12, 6))

        for i, color in enumerate(color_labels):
            hist = cv2.calcHist([img], [i], None, [256], [0, 256])
            color_histograms.append(hist)
            plt.subplot(1, 2, 1) 
            plt.plot(hist, color=color)
            plt.xlim([0, 256])
            plt.title("Color Image Histogram")
            plt.xlabel("Pixel Value")
            plt.ylabel("Pixel Count")

        gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        gray_hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

        plt.subplot(1, 2, 2)  
        plt.plot(gray_hist, color='gray')
        plt.xlim([0, 256])
        plt.title("Grayscale Image Histogram")
        plt.xlabel("Pixel Value")
        plt.ylabel("Pixel Count")

        plt.tight_layout() 

    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


image_path = "images/img.png"  
analyze_histogram(image_path)
```
